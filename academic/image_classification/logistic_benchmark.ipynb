```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Flatten, Dense, Activation
```

```python
train_dir = 'data/train/'
test_dir = 'data/test/'
```

```python
# number of output classes/categories
n_classes = 102

# specify batch size and number of epochs
batch_size = 64
n_epochs = 100

# create generator -------> COULD USE THIS INSTEAD OF MANUALLY RESIZING IMAGES!!!
datagen = ImageDataGenerator()

# prepare iterator for train/test sets
# use 'categorical' class mode for one-hot encoded y labels
train_it = datagen.flow_from_directory(train_dir, 
                                       target_size=(64,64),
                                       class_mode='categorical')
test_it = datagen.flow_from_directory(test_dir, 
                                      target_size=(64,64),
                                      class_mode='categorical')

# create a Sequential model (i.e. a sequential list of model layers)
model = Sequential()

# flatten image to 64x64x3 RGB image
model.add(Flatten(input_shape=(64,64,3)))

# add a fully-connected layer 
model.add(Dense(n_classes))

# add softmax activation output layer
model.add(Activation('softmax'))

# compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# fit the model
model.fit_generator(train_it,
                    steps_per_epoch=8200 // batch_size,
                    epochs=n_epochs)

# evaluate the model
loss = model.evaluate_generator(test_it, 
                                steps=880 // batch_size)

print('\nTest loss:', loss[0])
print('Test accuracy:', loss[1])
```
