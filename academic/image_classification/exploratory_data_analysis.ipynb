### Visualizations of number of images per category:

```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def count_images(data_dir):
    n_imgs = []
    
    for cat in os.listdir(data_dir):
        if cat != '.DS_Store':
            n_imgs.append(len(os.listdir(os.path.join(data_dir, cat))))
    
    return n_imgs
```

```python
def plot_dist(n_imgs_per_cat, title):
    plt.hist(n_imgs_per_cat)
    plt.title('Distribution of Number of Images per Category\n' + title)
    plt.xlabel('Number of Images')
    plt.ylabel('Number of Categories')
    plt.show()
```

```python
data_dir = 'data/101_ObjectCategories/'
train_dir = 'data/train/'
test_dir = 'data/test/'

n_imgs_per_cat = count_images(data_dir)
n_imgs_per_cat_train = count_images(train_dir)
n_imgs_per_cat_test = count_images(test_dir)
```

```python
plot_dist(n_imgs_per_cat, '(All Data)')
```

```python
plot_dist(n_imgs_per_cat_train, '(Training Set)')
```

```python
plot_dist(n_imgs_per_cat_test, '(Test Set)')
```

### Average number of images per category:

```python
def print_stats(n_imgs_per_cat, title):
    print('Average number of images per category ' + title + ': ' + str(np.mean(n_imgs_per_cat)))
    print('Median number of images per category ' + title + ': ' + str(np.median(n_imgs_per_cat)))
```

```python
print_stats(n_imgs_per_cat, '(all data)')
```

```python
print_stats(n_imgs_per_cat_train, '(training set)')
```

```python
print_stats(n_imgs_per_cat_test, '(test set)')
```
