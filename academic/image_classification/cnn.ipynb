```python
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.resnet50 import ResNet50
from keras.models import Model
from keras.layers import Flatten, Dense, Activation
```

```python
train_dir = 'data/train/'
test_dir = 'data/test/'
```

```python
# number of output classes/categories
n_classes = 102

# specify batch size and number of epochs
batch_size = 64
n_epochs = 100

# create generator -------> COULD USE THIS INSTEAD OF MANUALLY RESIZING IMAGES!!!
datagen = ImageDataGenerator()

# prepare iterator for train/test sets
# use 'categorical' class mode for one-hot encoded y labels
train_it = datagen.flow_from_directory(train_dir, 
                                       target_size=(64,64),
                                       class_mode='categorical')
test_it = datagen.flow_from_directory(test_dir, 
                                      target_size=(64,64),
                                      class_mode='categorical')

# load ResNet50 model without classification layer 
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(64,64,3))

# add new layers
resnet_output = base_model.output
new_flat = Flatten(input_shape=(64,64,3))(resnet_output)
new_class = Dense(1024)(new_flat) 
new_activation = Activation('relu')(new_class)
output_layer = Dense(n_classes)(new_activation) 
output_activation = Activation('softmax')(output_layer)

# define new model
model = Model(inputs=base_model.input, outputs=output_activation)

# freeze layers of base_model
for layer in base_model.layers:
    layer.trainable = False

# compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# fit the model
model.fit_generator(train_it,
                    steps_per_epoch=8200 // batch_size,
                    epochs=n_epochs)

# evaluate the model
loss = model.evaluate_generator(test_it, 
                                steps=880 // batch_size)

print('\nTest loss:', loss[0])
print('Test accuracy:', loss[1])
```
